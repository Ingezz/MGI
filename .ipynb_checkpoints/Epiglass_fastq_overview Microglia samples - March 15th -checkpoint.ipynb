```python
########
#Isolation 
cond1 = overview_fastq['archiveLoc'].str.contains('DGO') | overview_fastq['archiveLoc'].str.contains('DG') \
| overview_fastq['archiveLoc'].str.contains('JOS') | overview_fastq['archiveLoc'].str.contains('JoS') \
| overview_fastq['archiveLoc'].str.contains('AN') \
| overview_fastq['archiveLoc'].str.contains('SD') 

##Cell type /tissue 
cond2 = overview_fastq['archiveLoc'].str.contains('Microglia') | overview_fastq['archiveLoc'].str.contains('microglia') \
| overview_fastq['archiveLoc'].str.contains('WholeBrain') | overview_fastq['archiveLoc'].str.contains('wholebrain') \
| overview_fastq['archiveLoc'].str.contains('Whole_Brain') | overview_fastq['archiveLoc'].str.contains('brain')  \
| overview_fastq['archiveLoc'].str.contains('CD45') | overview_fastq['archiveLoc'].str.contains('CD11B')  \
| overview_fastq['archiveLoc'].str.contains('Monocytes') | overview_fastq['archiveLoc'].str.contains('monocytes') 

##Human only 
cond3 = overview_fastq['archiveLoc'].str.contains('Human') 

##Exclusion criteria 
cond4 = ~overview_fastq['archiveLoc'].str.contains('IPS')  & ~overview_fastq['archiveLoc'].str.contains('derived') \
& ~overview_fastq['archiveLoc'].str.contains('Breast') 

sum(cond1 & cond2 & cond3& cond4)
sOverview = overview_fastq[cond1 & cond2 & cond3 &cond4]

#sOverview.index
sOverview = sOverview[['sID','archiveLoc']]
sOverview.shape
```

```python
#############################Patient-ID 
patients = []
for rn in sOverview["sID"]:
    patient_match = re.match('\S*?\-P(?P<patient>[0-9]{1,2})\-\S*', rn)
    pat_match = re.match('\S*?\-Pat(?P<patient>[0-9]{1,2})\-\S*', rn)
    
    if patient_match:
#         
        patients.append("P" + str(int(patient_match.group('patient'))))

    elif pat_match:
        patients.append("P" + str(int(pat_match.group('patient'))))
    
    else:
        patients.append(0)
sOverview['patient'] = patients

print(sOverview)
```

```python
#############################Assay 
#####
assay = []
for rn in sOverview["sID"]:
     
    if re.search('RiboZero*', rn):
        assay.append("RNA_exp_RiboZero")        
        
    elif re.search('RNA*', rn):
        assay.append("RNA_exp")
        
    elif re.search('H3K27ac*', rn):
        assay.append("H3K27ac")
        
    elif re.search('H3K4Me2*', rn):
        assay.append("H3K4me2")
        
    elif re.search('H3K4me2*', rn):
        assay.append("H3K4me2")
        
    elif re.search('Input*', rn):
        assay.append("Input")
        
    elif re.search('PU1', rn):
        assay.append("PU.1")
        
    elif re.search('PU.1', rn):
        assay.append("PU.1")
        
    elif re.search('Sall1', rn):
        assay.append("Sall1")
        
    elif re.search('ATAC*', rn):
        assay.append("ATAC")
        
    else:
        assay.append("Na")
        
sOverview['assay'] = assay

print(sOverview)
```

```python
#############################Cell-type  
celltype = []
for rn in sOverview["sID"]:
     
    if re.search('Microglia*', rn): 
        if re.search('Culture*', rn):   
            celltype.append("MGin")                       
            
        else:
            celltype.append("MGex")

    elif re.search('Monocytes*', rn):
            celltype.append("Monocytes")   
            
    elif re.search('WholeBrain*', rn):
            celltype.append("WB")   
            
    elif re.search('CD45HighCD11bNegative-*', rn):
            celltype.append("CD45HighCD11bNegative-") 
            
    elif re.search('CD45HighCD11bNegative*', rn):
            celltype.append("CD45HighCD11bNegative")  

    elif re.search('CD11bCD45High*', rn):
            celltype.append("CD11b&CD45High")  
            
    elif re.search('CD45High*', rn):
            celltype.append("CD45High")    
    
    else: 
        celltype.append("Na")
    
sOverview['celltype'] = celltype

print(sOverview)
```

```python
######################Brain region 
##Three step : first step is identify the patients ID in sample info 
#Second step: Identify all samples for only one region is harvested and assign brain regions to matrix
# Third step: identify which samples from multy region samples belong to which 
##Or Maybe swithc up 2 and 3 

##STEP 1 
sample_info = pandas.read_csv("/home/iholtman/Human_natural_genetic_variation/March_10_2017_sample_info.csv")
sample_info = sample_info[sample_info['tissue_type'] == "Microglia"] ##Select microglia samples only 

patients = []

for rn in sample_info["Patient"]:
    patients.append("P" + str(int(rn[3:6])))
        
sample_info['patient_ID'] = patients

print(sample_info)
sample_info.shape
    
```

```python
##STEP 2 
sOverview['Brain_region'] = "Unknown"

for patient in sample_info["patient_ID"]:
    #print(patients.count(patient))
    #print(patient)
    
    if patients.count(patient) == 1:
        sOverview['Brain_region'][sOverview['patient'] == patient] = sample_info["Region"][sample_info["patient_ID"] == patient].iloc[0]
        
print(sOverview)

```

```python
###Brain region###
##STEP 3 
#for patient in sample_info["patient_ID"]:
for patient in sample_info["patient_ID"]:
    #print(patients.count(patient))
    
    if patients.count(patient) > 1:
        #print(sOverview['Brain_region'][sOverview['patient'] == patient])
        print(sOverview[sOverview['patient'] == patient])
```

```python
##STEP 3 
#for patient in sample_info["patient_ID"]:

for patient in sample_info["patient_ID"]:
    
    if patients.count(patient) > 1:
        
        brain_regions = [] 
    
        for sample_ID in sOverview['sID'][sOverview['patient'] == patient]:
                      
            if re.search('Occip*', sample_ID):
                brain_regions.append("Occipital")
                
            elif re.search('Tempor*', sample_ID):
                brain_regions.append("Temporal")

            elif re.search('Front*', sample_ID):
                brain_regions.append("Frontal")
                
            elif re.search('Pariet*', sample_ID):
                brain_regions.append("Parietal")
                
            elif re.search('Hippocampus*', sample_ID):
                brain_regions.append("Hippocampus")
                         
            elif re.search('Culture*', sample_ID):
                brain_regions.append("Culture")
                
            elif re.search('Occ-Par*', sample_ID):
                brain_regions.append("Occ-Par")
                
            elif re.search('Medulla*', sample_ID):
                brain_regions.append("Medulla")

            elif re.search('Tumor*', sample_ID):
                brain_regions.append("Tumor")
                
            elif re.search('Mix*', sample_ID):
                brain_regions.append("Mixed")
                
            else:        
                brain_regions.append("Unknown")
                
        sOverview['Brain_region'].loc[sOverview['patient'] == patient] = brain_regions 
       
print(sOverview[sOverview['Brain_region'] == "Unknown"])
            

```

```python
##
Condition = []

for index, row in sOverview.iterrows():
    if row["Condition_tissue"] == "Tumor":
        Condition.append("Tumor")
        
    else: 
        if patient_info["Broad_Diagnosis"].loc[patient_info['patient_ID'] == row["patient"]].iloc[0] == "Epilepsy":
            Condition.append("Epileptic")

        elif patient_info["Broad_Diagnosis"].loc[patient_info['patient_ID'] == row["patient"]].iloc[0] == "Tumor":
            Condition.append("Tumor margin tissue")
            
        elif patient_info["Broad_Diagnosis"].loc[patient_info['patient_ID'] == row["patient"]].iloc[0] == "Acute":
            Condition.append("Acute")

        else:
            Condition.append("Unknown")


sOverview["Condition_tissue"] = Condition
print(sOverview)
```

```python
###################
#Isolation performed by  
sOverview["Isolator"] = "not_evaluated"

Isolator = []

for rn in sOverview["sID"]:
   
    if re.search('DGO*', rn):
        Isolator.append("DGO")
                
    elif re.search('DG*', rn):
        Isolator.append("DGO")
        
    elif re.search('JOS*', rn):
        Isolator.append("JOS")

    else:
        Isolator.append("Na")
        

sOverview["Isolator"] = Isolator

print(sOverview)


```

```python
###################
#Quality flag - add 
sOverview["Quality_flag"] = "not_evaluated"

evaluation = []

for rn in sOverview["archiveLoc"]:
    evaluate = re.search("Invalid", rn)
    
    if evaluate:  
        evaluation.append("Invalid")
    else:
        evaluation.append("not_evaluated")

sOverview["Quality_flag"] = evaluation

print(sOverview)
```

```python
import numpy 
sOverview.index = sOverview["sID"]
sOverview.sort_index(inplace=True)

sOverview['in_vitro'] = [i.lower().find('culture') > -1 for i in sOverview.index]
sOverview['tgfb'] = [i.lower().find('tgf') > -1 for i in sOverview.index]
sOverview['csf1'] = [i.lower().find('csf1') > -1 or i.lower().find('csf1') > -1 for i in sOverview.index]
sOverview['mcsf'] = [i.lower().find('mcsf') > -1 or i.lower().find('m_csf') > -1 for i in sOverview.index]
sOverview['il34'] = [i.lower().find('il34') > -1 for i in sOverview.index]
sOverview['7d'] = ['6d' in i.lower() or '7d' in i.lower() or '8d' in i.lower() for i in sOverview.index]
sOverview['10d'] = [i.lower().find('10d') > -1 for i in sOverview.index]
sOverview['6h'] = ['6h' in i.lower() or '7h' in i.lower() for i in sOverview.index]
sOverview['24h'] = [i.lower().find('24h') > -1 or i.lower().find('1d') > -1 for i in sOverview.index]
sOverview['72h'] = [i.lower().find('72h') > -1 for i in sOverview.index]
sOverview['neural_media'] = [i.lower().find('NeuralMedia'.lower()) > -1 for i in sOverview.index]
sOverview['human_serum'] = ['humanserum'.lower() in i.lower() for i in sOverview.index]
sOverview['stempro'] = ['stempro'.lower() in i.lower() for i in sOverview.index]
sOverview['isoproterenol'] = ['Isoproterenol'.lower() in i.lower() for i in sOverview.index]
sOverview['dmso'] = ['dmso'.lower() in i.lower() for i in sOverview.index]
sOverview['tgfbr1antagonist'] = ['TGFBR1antagonist'.lower() in i.lower() for i in sOverview.index]
sOverview['ribozero'] = [i.find('RiboZero') > -1 for i in sOverview.index]
sOverview['fetal'] = [i.lower().find('fetal') > -1 for i in sOverview.index]
sOverview['lps'] = [i.lower().find('lps') > -1 for i in sOverview.index]


#######
stimulations_list = []

for index, row in sOverview.iterrows():
    stimulations = []
        
    if row['7d'] == True:
        stimulations.append("7d")
        
    if row['10d'] == True:
        stimulations.append("10d")

    if row['6h'] == True:
        stimulations.append("6h")
         
    if row['24h'] == True:
        stimulations.append("24h")
        
    if row['72h'] == True:
        stimulations.append("72h")

    if row['tgfb'] == True:
        stimulations.append("tgfb")
        
    if row['mcsf'] == True:
        stimulations.append("mcsf")
        
    if row['csf1'] == True:
        stimulations.append("csf1")

    if row['il34'] == True:
        stimulations.append("il34")
        
    if row['neural_media'] == True:
        stimulations.append("neural_media")
       
    if row['human_serum'] == True:
        stimulations.append("human_serum")
        
    if row['stempro'] == True:
        stimulations.append("stempro")

    if row['isoproterenol'] == True:
        stimulations.append("isoproterenol")
         
    if row['dmso'] == True:
        stimulations.append("dmso")
        
    if row['tgfbr1antagonist'] == True:
        stimulations.append("tgfbr1anta")

    if row['fetal'] == True:
        stimulations.append("fetal")

    if row['lps'] == True:
        stimulations.append("lps")
        
    stimulations_list.append(" + ".join(stimulations))

sOverview["Culture_conditions"] = stimulations_list

del sOverview['in_vitro'] 
del sOverview['tgfb'] 
del sOverview['mcsf'] 
del sOverview['csf1'] 
del sOverview['il34'] 
del sOverview['7d'] 
del sOverview['10d'] 
del sOverview['6h'] 
del sOverview['24h'] 
del sOverview['72h'] 
del sOverview['neural_media'] 
del sOverview['human_serum'] 
del sOverview['stempro'] 
del sOverview['isoproterenol'] 
del sOverview['dmso'] 
del sOverview['tgfbr1antagonist'] 
del sOverview['ribozero'] 
del sOverview['fetal'] 
del sOverview['lps'] 
del sOverview['sID'] 
```
