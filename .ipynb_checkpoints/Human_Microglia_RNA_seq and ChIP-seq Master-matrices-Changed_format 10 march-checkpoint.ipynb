```python
############################
########Location of fastq-files used for genotyping and add patientnumber as a separate column  
import glob
import os
import pandas
import re

dir=  "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/5-dec/"
dir2= "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/cat_fastq_files_28_nov/"

files= glob.glob(dir +"*.fastq.gz") + glob.glob(dir2 +"*.fastq.gz")

folders = pandas.DataFrame([], index=files)
folders.index.name='fastq'

folders.sort_index(inplace=True)

folders['cat'] = [i.lower().find('cat') > -1 for i in folders.index]

#####
patients = []
for rn in folders.index:
    patient_match = re.match('\S*?\_P(?P<patient>[0-9]{1,2})\S*', rn)
    if patient_match:
#         
        patients.append(int(patient_match.group('patient')))
    else:
        patients.append(0)
folders['patient'] = patients

patientz = []
for rn in folders.index:
    patient_match = re.match('\S*?\-P(?P<patient>[0-9]{1,2})\-\S*', rn)
    if patient_match:   
        patientz.append(int(patient_match.group('patient')))
    else:
        patientz.append(0)
    
folders['patientz'] = patientz

folders["patient"] = folders["patient"] + folders["patientz"]
#print(folders["patient"])
del folders['patientz']

print(folders[1:5])
```

```python
############################
########Location of fastq-files used for genotyping and add patientnumber as a separate column  
import glob
import os
import pandas
import re

dir=  "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/5-dec/"
dir2= "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/cat_fastq_files_28_nov/"

files= glob.glob(dir +"*.fastq.gz") + glob.glob(dir2 +"*.fastq.gz")

folders = pandas.DataFrame([], index=files)
folders.index.name='fastq'

folders.sort_index(inplace=True)

folders['cat'] = [i.lower().find('cat') > -1 for i in folders.index]

#####
patients = []
for rn in folders.index:
    patient_match = re.match('\S*?\_P(?P<patient>[0-9]{1,2})\S*', rn)
    if patient_match:
#         
        patients.append(int(patient_match.group('patient')))
    else:
        patients.append(0)
folders['patient'] = patients

patientz = []
for rn in folders.index:
    patient_match = re.match('\S*?\-P(?P<patient>[0-9]{1,2})\-\S*', rn)
    if patient_match:   
        patientz.append(int(patient_match.group('patient')))
    else:
        patientz.append(0)
    
folders['patientz'] = patientz

folders["patient"] = folders["patient"] + folders["patientz"]
#print(folders["patient"])
del folders['patientz']

print(folders[1:5])
```

```python
############################
########Location of fastq-files used for genotyping and add patientnumber as a separate column  
import glob
import os
import re

dir=  "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/5-dec/"
dir2= "/data/scratch/iholtman/human_microglia_genotyping/fastq-files/fastq-genotyping/cat_fastq_files_28_nov/"

files= glob.glob(dir +"*.fastq.gz") + glob.glob(dir2 +"*.fastq.gz")

files_genotyping_RNA = pandas.DataFrame(index=files)
files_genotyping_RNA.index.name='fastq'
files_genotyping_RNA.sort_index(inplace=True)

files_genotyping_RNA['cat'] = [i.lower().find('cat') > -1 for i in files_genotyping_RNA.index]

#####
patients = []
for rn in files_genotyping_RNA.index:
    patient_match = re.match('\S*?\_P(?P<patient>[0-9]{1,2})\S*', rn)
    if patient_match:
#         
        patients.append(int(patient_match.group('patient')))
    else:
        patients.append(0)
files_genotyping_RNA['patient'] = patients

patientz = []
for rn in files_genotyping_RNA.index:
    patient_match = re.match('\S*?\-P(?P<patient>[0-9]{1,2})\-\S*', rn)
    if patient_match:   
        patientz.append(int(patient_match.group('patient')))
    else:
        patientz.append(0)
    
files_genotyping_RNA['patientz'] = patientz

files_genotyping_RNA["patient"] = files_genotyping_RNA["patient"] + files_genotyping_RNA["patientz"]

del files_genotyping_RNA["patientz"]

##
patient_id = []

for patient in files_genotyping_RNA["patient"]:

    if patient < 10:
        patient_id.append("HMG000"+ str(patient))
    else:
        patient_id.append("HMG00"+ str(patient))
        
files_genotyping_RNA["Patient_ID"] = patient_id

del files_genotyping_RNA["patient"]

files_genotyping_RNA["assay"] = "RNA-seq_genotyping"

 
###
files_genotyping_RNA["Info"] = "Na"
concat = files_genotyping_RNA["Info"][:]

concat[files_genotyping_RNA["cat"] == True] = "Concatenation"
concat[files_genotyping_RNA["cat"] == False] = "single_fastq"

files_genotyping_RNA["Info"] = concat
del files_genotyping_RNA["cat"]

#####
files_genotyping_RNA["fastq_path"] = files_genotyping_RNA.index[:] 
files_genotyping_RNA["Tissue"] = "unknown"
files_genotyping_RNA.index = range(100, 123)

print(files_genotyping_RNA)
```
