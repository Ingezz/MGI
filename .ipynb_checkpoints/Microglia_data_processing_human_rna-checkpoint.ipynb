```python
all_folders = pandas.concat((invitro_folders, exvivo_folders, brain_folders), axis=0)
all_folders.sort_index(inplace=True)

all_folders['tgfb'] = [i.lower().find('tgf') > -1 for i in all_folders.index]
all_folders['mcsf'] = [i.lower().find('mcsf') > -1 or i.lower().find('m_csf') > -1 for i in all_folders.index]
all_folders['il34'] = [i.lower().find('il34') > -1 for i in all_folders.index]
all_folders['7d'] = ['6d' in i.lower() or '7d' in i.lower() or '8d' in i.lower() for i in all_folders.index]
all_folders['10d'] = [i.lower().find('10d') > -1 for i in all_folders.index]
all_folders['6h'] = ['6h' in i.lower() or '7h' in i.lower() for i in all_folders.index]
all_folders['24h'] = [i.lower().find('24h') > -1 or i.lower().find('1d') > -1 for i in all_folders.index]
all_folders['72h'] = [i.lower().find('72h') > -1 for i in all_folders.index]
all_folders['neural_media'] = [i.lower().find('NeuralMedia'.lower()) > -1 for i in all_folders.index]
all_folders['human_serum'] = ['humanserum'.lower() in i.lower() for i in all_folders.index]
all_folders['stempro'] = ['stempro'.lower() in i.lower() for i in all_folders.index]
all_folders['isoproterenol'] = ['Isoproterenol'.lower() in i.lower() for i in all_folders.index]
all_folders['dmso'] = ['dmso'.lower() in i.lower() for i in all_folders.index]
all_folders['wholebrain'] = [not numpy.isnan(all_folders.loc[i, 'wholebrain']) for i in all_folders.index]
all_folders['tgfbr1antagonist'] = ['TGFBR1antagonist'.lower() in i.lower() for i in all_folders.index]

all_folders['ribozero'] = [i.find('RiboZero') > -1 for i in all_folders.index]
all_folders['use'] = [sum([f.lower().find(exclude_term.lower()) > -1 for exclude_term in use_disqualifiers]) == 0 for f in all_folders.index]
# all_folders['primary'] = [sum([f.lower().find(exclude_term.lower()) > -1 for exclude_term in primary_disqualifiers]) == 0 for f in all_folders.index]
# all_folders['primary'] = numpy.logical_and(all_folders['primary'], all_folders['use'])
all_folders['primary'] = [rep in PAPER_SAMPLES for rep in all_folders.index]
all_folders['fetal'] = [i.lower().find('fetal') > -1 for i in all_folders.index]
all_folders['lps'] = [i.lower().find('lps') > -1 for i in all_folders.index]
all_folders['0h'] = ~all_folders.in_vitro & all_folders.primary

patients = []
for rn in all_folders.index:
    patient_match = re.match('\S*?\_P(?P<patient>[0-9]{1,2})\_\S*', rn)
    if patient_match:
#         
        patients.append(int(patient_match.group('patient')))
    else:
        patients.append(0)
all_folders['patient'] = patients
        
# all_folders[all_folders['primary'] == True].index
```

# Find and copy / rename fastq files

```python
import os
import shutil
import subprocess

fastq_mapping = pandas.read_csv('/home/dskola/human/dbgap_submission/human_rna_tag_fastq_mapping.csv', index_col=0).fastq_path.to_dict()
DESTINATION_FOLDER='/home/dskola/human/dbgap_submission/submission_folder/'

missing_samples=set(name_mapping.keys()).difference(fastq_mapping.keys())
print('\nSamples missing FASTQ paths:\n')
print('\n'.join(missing_samples))
print('\n')
open('/home/dskola/human/dbgap_submission/missing_rna.txt', 'wt').write('\n'.join(missing_samples))
for tag_name, simple_name in name_mapping.items():
    if tag_name in fastq_mapping:        
        fastq_path = fastq_mapping[tag_name]
        print('\tProcessing sample {} ({})'.format(tag_name, simple_name))
        
        destination_path = os.path.join(DESTINATION_FOLDER, '{}.fastq.gz'.format(simple_name))
        
        if fastq_path.endswith('.fastq.gz'):
            print('\t\tSample appears to be a single compressed file')
            need_copy = True
            if os.path.isfile(destination_path):
                source_size = os.stat(fastq_path)[6]
                dest_size = os.stat(destination_path)[6]
                if source_size == dest_size:
                    need_copy = False
            if need_copy:
                print('\t\tCopying...', end="", flush=True)
                shutil.copyfile(fastq_path, destination_path)
                print('done.')
            else:
                print('\t\tNo need to copy')
        elif fastq_path.endswith('.fastq'):
            print('\t\tSample appears to be a single uncompressed file')
            need_copy = True
            if os.path.isfile(destination_path):
                need_copy = False
            if need_copy:
                cmd_line = ['cat', fastq_path, '|', 'gzip', '-c', '-9', '>', destination_path]
                print('\t\tCopying and gzipping...', end="", flush=True)
                output = subprocess.check_output(' '.join(cmd_line), shell=True, stderr=subprocess.STDOUT)
                print('done.')
            else:
                print('\t\tNo need to copy')
        else:
            # concatenate all subfiles
            subfiles = [os.path.join(fastq_path,f) for f in os.listdir(fastq_path) if f.endswith('.fastq.gz') or f.endswith('.fastq')]
            if len(subfiles) > 0:
                print('\t\tSample appears to have {} lane files'.format(len(subfiles)))
                need_copy = True
            else:
                raise Exception('No lane subfiles found in path:\n{}'.format(fastq_path))
            if os.path.isfile(destination_path):
                dest_size = os.stat(destination_path)[6]
                if dest_size > 10000:
                    need_copy = False
            if need_copy:
                destination_tmp_file = destination_path + '.tmp'
                for fname in subfiles:
                    
                    if fname.endswith('.fastq'):
                        cmd_line = ['cat', fname, '>>', destination_tmp_file]
                    elif fname.endswith('.fastq.gz'):
                        cmd_line = ['zcat', fname, '>>', destination_tmp_file]
#                     print(' '.join(cmd_line))
                    print('\t\tConcatenating lane file {} ...'.format(fname), end="", flush=True)

                    output = subprocess.check_output(' '.join(cmd_line), shell=True, stderr=subprocess.STDOUT)
                    print('done.')

                cmd_line = ['cat', destination_tmp_file, '|', 'gzip', '-c', '-9', '>', destination_path]                
#                 print(' '.join(cmd_line))

                print('\t\tGzipping concatenated file...', end="", flush=True)               

                output = subprocess.check_output(' '.join(cmd_line), shell=True, stderr=subprocess.STDOUT)
                os.remove(destination_tmp_file)
                print('done.')
            else:
                print('\t\tNo need to copy')
        
    else:
        print('\n!!No FASTQ path found for sample {} ({})!!\n'.format(tag_name, simple_name))
```
