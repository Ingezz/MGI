```python
##### convert to pairwise matrix 
######Combine the picard concordance contingency tables in one matrix 
import glob
import os
import pandas
files = glob.glob("/data/scratch/iholtman/human_microglia_genotyping/compare_imputation_WGS/concordance_picard/*.vcf.genotype_concordance_contingency_metrics")

#data = pandas.DataFrame(pandas.read_csv(file, skiprows=6, skipfooter= 7, sep= "\t"))
data = pandas.DataFrame()

for file in files:
   data = data.append(pandas.read_csv(file, skiprows=6, sep= "\t"))

data = data[data.VARIANT_TYPE == "SNP"]

data['True'] = data['TP_COUNT'] + data['TN_COUNT'] 
data['False'] = data['FP_COUNT'] + data['FN_COUNT'] 

data['ratio_T/F'] = data['True'] / data['False']
data = data.sort_values(by ='ratio_T/F', ascending = False)

data = data.sort_values(by =['TRUTH_SAMPLE', 'CALL_SAMPLE'], )
print(data.shape)

print(data[1:10])   
```

```python
pairwise = pandas.DataFrame(index = set(data['TRUTH_SAMPLE']), columns = set(data['CALL_SAMPLE']))

counter = 0 
ratios = data['ratio_T/F']

for truth in set(data['TRUTH_SAMPLE']):
    for call in set(data['CALL_SAMPLE']):        
        pairwise.set_value(truth, call, ratios.iloc[counter])
        counter = counter + 1
        
print(pairwise)

```
